```python
avg_Schizophrenia_per_year = data.groupby('Year')['Schizophrenia'].mean().reset_index()
avg_Schizophrenia_per_year
```

```python
plt.figure(figsize=(12,5))
sb.lineplot(data = avg_Schizophrenia_per_year, x = "Year", y = "Schizophrenia", marker = "o")
plt.title("Average Schizophrenia per year")
plt.xlabel("Year")
plt.ylabel("Schizophrenia")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sb.barplot(data = avg_Schizophrenia_per_year, x = "Year", y = "Schizophrenia")
plt.title("Average Schizophrenia per year")
plt.xlabel("Year")
plt.ylabel("Schizophrenia")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_Schizophrenia_per_year = avg_Schizophrenia_per_year.sort_values(by="Schizophrenia", ascending=True)
plt.figure(figsize=(14, 6))
sb.barplot(data=avg_Schizophrenia_per_year, x="Year", y="Schizophrenia", palette="viridis")

plt.title("Average Schizophrenia by Year (Sorted from Lowest to Highest)", fontsize=14)
plt.xlabel("Year (Sorted by Schizophrenia)", fontsize=12)
plt.ylabel("Average Schizophrenia", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')

plt.show()
```

```python
min_Schizophrenia = avg_Schizophrenia_per_year["Schizophrenia"].min()
max_Schizophrenia = avg_Schizophrenia_per_year["Schizophrenia"].max()

plt.figure(figsize=(14, 6))
sb.barplot(data=avg_Schizophrenia_per_year, x="Year", y="Schizophrenia", palette="viridis")

plt.ylim(min_Schizophrenia*0.98, max_Schizophrenia*1.02)
plt.title("Average Schizophrenia by Year (Sorted from Lowest to Highest)", fontsize=14)
plt.xlabel("Year (Sorted by Schizophrenia)", fontsize=12)
plt.ylabel("Average Schizophrenia", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')

plt.show()
```
